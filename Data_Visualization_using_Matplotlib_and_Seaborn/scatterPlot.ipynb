```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
colors = ['red' if score < 60 else 'green' for score in exam_scores]
size = [2 * score for score in exam_scores]
Social_score = [50, 55, 60, 65, 70, 75, 80, 85, 90]
plt.scatter(study_hours, exam_scores, c=Social_score, marker='o', s=size, alpha=0.7, edgecolors='black', cmap='viridis')
plt.colorbar(label='Social Scores')
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.grid(True)
plt.show()
```

```python
plt.scatter(study_hours, Social_score, c=exam_scores, marker='s', s=size, alpha=0.7, edgecolors='black', cmap='plasma')
plt.colorbar(label='Exam Scores')
plt.title("Study Hours vs Social Scores")
plt.xlabel("Study Hours")
plt.ylabel("Social Scores")
for i in range(len(study_hours)):
    plt.text(study_hours[i], Social_score[i], f'({study_hours[i]},{Social_score[i]})')
plt.grid(True)
plt.show()
```

```python
plt.scatter(study_hours, Social_score, c=exam_scores, marker='s', s=size, alpha=0.7, edgecolors='black', cmap='plasma')
plt.colorbar(label='Exam Scores')
plt.title("Study Hours vs Social Scores")
plt.xlabel("Study Hours")
plt.ylabel("Social Scores")
for i in range(len(study_hours)):
    plt.annotate(f'S{i + 1}', (study_hours[i], Social_score[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.grid(True)
plt.show()
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, color='blue', marker='o', s=100, label='Class A', alpha=0.7, edgecolors='black')
plt.scatter(class_b_hours, class_b_scores, color='orange', marker='s', s=100, label='Class B', alpha=0.7, edgecolors='black')
plt.title("Study Hours vs Exam Scores by Class")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.legend()
plt.grid(True)
plt.show()
```
