```python
import matplotlib.pyplot as plt
```

```python
activity = ['Sleeping', 'Eating', 'Working', 'Playing', 'Learning']
timespent = [7, 2, 8, 3, 4]
color = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']
plt.figure(figsize=(6,8))
plt.pie(timespent, labels=activity, colors=color, autopct='%1.1f%%', startangle=140)
plt.title("Daily Activity Distribution")
plt.axis('equal')
plt.show()
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

lables = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.figure(figsize=(6,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels= lables, colors=color, alpha = 1)
# plt.legend(loc = 'upper left')
plt.legend(loc = 'best')
plt.title("Student time distribution")
plt.xlabel("Days of the week")
plt.ylabel("Hours")
plt.xticks(days)
plt.grid(True)
plt.show()
```
