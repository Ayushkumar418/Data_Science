```python
tips = sns.load_dataset("tips")
print(tips.head())
print(tips.info())
```

```python
# sns.lineplot(data = tips, x="total_bill", y="tip", hue="smoker")
sns.scatterplot(data = tips, x="total_bill", y="tip", )
plt.grid()
```

```python
flihgts = sns.load_dataset('flights')
print(flihgts.head())
pv = flihgts.pivot(index = 'month', columns='year', values='passengers')
print(pv)
```

```python
sns.heatmap(pv, annot=True, fmt="d", cmap="jet")
```

```python
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.show()
```
