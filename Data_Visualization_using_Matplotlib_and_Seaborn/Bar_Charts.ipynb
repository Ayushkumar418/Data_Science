```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
```

```python
x = np.arange(len(years))
width = 0.25
```

```python
plt.bar(x - width, sachin, width=width, color = 'blue', label = "Sachin")
plt.bar(x, sehwag, width=width, color = 'green', label = "Sehwag")
plt.bar(x + width, kohli, width=width, color = 'orange', label = "Kohli")

plt.xlabel("Years --->")
plt.ylabel("Runs Scored --->")
plt.xticks(x, years)
plt.title("Runs scored by Batsmen over the Years")
plt.legend()
plt.tight_layout()
```
