```python
import matplotlib.pyplot as plt
```

```python
data = [2, 15, 13, 7, 19, 21, 22, 18, 16, 4, 20, 35, 35, 28, 27, 26, 24, 23, 29, 31, 35, 33, 32, 34, 36, 38, 40, 39, 37, 41, 42, 45, 44, 43, 76, 48, 47, 49, 50, 52, 53, 55, 54, 51, 56, 58, 57, 59, 60, 65]
print(len(data))
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.hist(data, bins=5, edgecolor = 'black', color="red", alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60]
plt.hist(ages, bins=bins, edgecolor='black', color='skyblue', alpha=0.7)
plt.title("Age Distribution of Participants")
plt.xlabel("Age Groups")
plt.ylabel("Number of Participants")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.hist(ages, bins=bins, edgecolor='black', color='skyblue', alpha=0.7)
plt.axvline(x=30, color='red', linestyle='dashed', linewidth=1, label='Age 30')
plt.axvline(x=35, color='blue', linestyle='dashed', linewidth=1, label='Age 35')
plt.legend()
plt.title("Age Distribution of Participants")
plt.xlabel("Age Groups")
plt.ylabel("Number of Participants")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
