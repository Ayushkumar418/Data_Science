```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
# plt.pie(runs, labels=labels, autopct='%1.1f%%', startangle=140)
explode = (0.1, 0, 0.05, 0)  # only "explode" the 1st slice (i.e. 'Sachin')
plt.pie(runs, labels=labels, wedgeprops = {'edgecolor':'black', 'linewidth':2, 'linestyle': "-"}, autopct='%1.2f%%', startangle=120, explode = explode)
plt.title("Runs scored by players in International Cricket")
plt.show()
```

```python
# plt.pie(runs, labels=labels, autopct='%1.1f%%', startangle=140)
explode = (0.1, 0, 0.05, 0)  # only "explode" the 1st slice (i.e. 'Sachin')
plt.pie(runs, labels=labels, wedgeprops = {'edgecolor':'black', 'linewidth':2, 'linestyle': "-"}, autopct='%1.2f%%', startangle=120, explode = explode, shadow=True)
plt.title("Runs scored by players in International Cricket")
plt.show()
```
