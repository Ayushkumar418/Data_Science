```python
import pandas as pd
```

### Series

```python
s = pd.Series([40,60, 68, 72, 85, 93, 31, 34, 23, 45])
print(s)
print(type(s))
```

```python
s1 = pd.Series([29, 35, 46, 52, 61], index = ["Ayush", "Rohan", "Ankit", "Sagar", "Vikram"])
print(s1)
print(type(s1))
print(s1["Ayush"])
print(s1[0])
```

### DataFrame

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)
print(df)
```
