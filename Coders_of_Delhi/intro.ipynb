```python
import json
```

```python
# The function to load JSON data from a file
def load_data(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data
```

```python
data = load_data('data.json')
print(data)
```

```python
# Lets Write a function to users and there connections
def display_users(data):
    print("Users and their connections:")
    for user in data['users']:
        print(f"{user['id']}. {user["name"]} is friends with: {user["friends"]} and likes pages are {user["liked_pages"]}")

    print("\nPages Inforamation:")
    for page in data['pages']:
        print(f"page no. {page['id']} and community page of {page['name']}")
        
display_users(data)
```
