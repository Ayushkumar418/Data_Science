```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as f:
        return json.load(f)
```

```python
def find_pages_you_might_know(user_id, data):
    user_pages = {}
    for user in data['users']:
        user_pages[user['id']] = set(user['liked_pages'])
        
    if user_id not in user_pages:
        return []
    
    user_likes_pages = user_pages[user_id]
    page_suggestions = {}
    
    for other_user, pages in user_pages.items():
        if other_user == user_id:
            continue
        shared_pages = user_likes_pages.intersection(pages)
        for page in pages:
            if page not in user_likes_pages:
                page_suggestions[page] = page_suggestions.get(page, 0) + len(shared_pages)
                
    return [page_id for page_id, score in sorted(page_suggestions.items(), key=lambda item: item[1], reverse=True) if score >= 2]
```

```python
data = load_data('massive_data.json')
user_id = 1
suggestions = find_pages_you_might_know(user_id, data)
print(suggestions)
```
