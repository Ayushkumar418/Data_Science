```python
import json
```

```python
def clean_data(data):
    # Remove users with empty names
    data['users'] = [user for user in data['users'] if user['name'].strip()]
    
    # Remove duplicate friends in each user's friends list
    for user in data['users']:
        user['friends'] = list(set(user['friends']))
        
    data['users'] = [user for user in data['users'] if user['friends'] or user['liked_pages']]
    
    # Remove duplicate pages based on 'id'
    unique_pages = {}
    for page in data['pages']:
        unique_pages[page['id']] = page
    data['pages'] = list(unique_pages.values())
    
    return data
```

```python
data = json.load(open('data2.json'))
cleaned_data = clean_data(data)
json.dump(cleaned_data, open('cleaned_data2.json', 'w'), indent=4)
print("Data cleaning complete. Cleaned data saved to 'cleaned_data2.json'.")
```
